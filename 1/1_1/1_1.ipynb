## SỬ DỤNG KỸ THUÂT TÁCH CHUỖI - SPLIT

```python
diemthi = df1.loc[0, 'DIEM_THI']
diemthi
```

```python
# Toán:   2.00   Ngữ văn:   5.50   Lịch sử:   3.00   Địa lí:   5.00
lst_diemthi = diemthi.replace(':', '').split('   ')
lst_diemthi
```

```python
# lst_mon = ['Toán', 'Ngữ văn', 'Địa lí', 'Lịch sử', 'Tiếng Anh', 'Sinh học', 'Vật lí', 'Hóa học']
```

```python
def vi_tri_diem_mon_hoc( lst_mon, lst_diemthi):
    if lst_mon in lst_diemthi:
        vitrimon = lst_diemthi.index(lst_mon)
        diem = lst_diemthi[vitrimon + 1]
        return round(float(diem),2)
    else:
        return np.nan
```

```python
vi_tri_diem_mon_hoc('Tiếng Lào',lst_diemthi)
```

```python
df1['DS DIEM THI'] = df1['DIEM_THI'].str.replace(':','').str.split("   ")
df1.head()
```

```python
df1['DS2 DIEM THI'] = df1['DIEM_THI'].apply(lambda x: x.replace(':','').split("   "))
df1.head()
```

```python
for mon in lst_mon:
    df1[mon] = df1['DS2 DIEM THI'].apply(lambda x: vi_tri_diem_mon_hoc(mon, x))
    
```

## SỬ DỤNG REGULAR EXPRESSION

```python
df2 = df.copy()
df2.head()
```

```python
lst_mon = ['Toán', 'Ngữ văn', 'Địa lí', 'Lịch sử', 'Tiếng Anh', 'Sinh học', 'Vật lí', 'Hóa học']
```

```python
# ReEx có parten dang: Toán:\s{3}\d+\.\d{2}
for tenmon in lst_mon:
    df2[tenmon] = df2['DIEM_THI'].str.extract('(' + tenmon + ':\s{3}\d+.\d{2})', expand = True)[0].str.split(":   ", expand = True)[1]
```
