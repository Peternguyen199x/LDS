```python
df1 = df.copy()
# cac gia tri cot PRICE khong phai number 
df1['price'].str.isnumeric().value_counts()  
df1.info()
```

```python
# thay the cac gia tri khong phai so cua PRICE = median(price_so)
# Xac dinh median(price_so)
price_so = df1['price'].loc[df1['price'].str.isnumeric() == True]
pmedian = price_so.median()
print(pmedian)
# thay the gia tri pmedian vao vi tri k phai so cua PRICE
price_not_number = df1['price'].loc[df1['price'].str.isnumeric() == False]
print(price_not_number)
df1['price']= df1['price'].replace(price_not_number.tolist(), pmedian).astype(float)
df1.info()
```
